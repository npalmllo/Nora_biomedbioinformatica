```python
# Ahora vamos a por un ejemplo un poco más complicado. Queremos que nos devuelva TODOS los numeros pares en un lista. 
# Si no hay números pares, nos va a devolver la lista vacía
def es_par(lista_numeros):
    # Creamos una lista vacía
    lista_pares = []
    # Parseamos cada elemento de nuestro input:
    for numero in lista_numeros:
        if numero % 2 == 0:
            # Si esto es verdad, añadimos nuestro elemento a la lista
            lista_pares.append(numero)
        else:
            pass
    return lista_pares
```

#### 1.- El menor de dos pares. Escribe una función que devuelva el menor de dos números, SOLAMENTE si los dos números son pares. En caso de que uno de los dos, o los dos, sean impares, devuelve el número más alto

```python
#creamos la funcion de el menor de dos pares, 
def menor_de_dos_pares(a, b):
    if a % 2 == 0 and b % 2 == 0:  # Ambos pares
        return min(a, b)
    else:  # Uno o ambos impares
        return max(a, b)
```

```python
# ahora ejecutamos los ejemplos de prueba:
print(menor_de_dos_pares(2, 4))   # Ambos pares → menor → 2
print(menor_de_dos_pares(2, 5))   # Uno impar → mayor → 5
print(menor_de_dos_pares(7, 9))   # Ambos impares → mayor → 9
print(menor_de_dos_pares(8, 3))   # Uno impar → mayor → 8
```

#### 2.- Escribe una función que tome como argumento una string de dos palabras y devuelva True si ambas empiezan por la misma letra

```python
def misma_letra(palabras):
    palabra1, palabra2 = palabras.split()  # separa las dos palabras
    return palabra1[0].lower() == palabra2[0].lower()
```

```python
# Ejemplos de prueba:
print(misma_letra("Hola Hermanos"))   # True (H y H)
print(misma_letra("Perro Gato"))      # False (P y G)
print(misma_letra("mundo maravilloso"))  # True (m y m)
```

#### 4.- Escribe una función que devuelva en mayúsculas la primera letra y la quinta letra de una palabra. Pista: usa capitalize()

```python
def mayus_primera_quinta(palabra):
    if len(palabra) < 5:
        return "La palabra debe tener al menos 5 letras"
    
    # Convertimos la palabra en lista de caracteres (porque las strings son inmutables)
    lista = list(palabra)
    lista[0] = lista[0].upper()
    lista[4] = lista[4].upper()
    
    return "".join(lista)
```

```python
print(mayus_primera_quinta("python"))     # Python -> 'PythoN'
print(mayus_primera_quinta("murcielago")) # MurcIelago
print(mayus_primera_quinta("casa"))       # Mensaje de error
```

#### 6.- Dada una lista de enteros, devuelve True si contiene un 3 al lado de otro 3

```python
def tiene_33(numeros):
    for i in range(len(numeros) - 1):  # -1 para no salirnos del rango
        if numeros[i] == 3 and numeros[i+1] == 3:
            return True
    return False
```

```python
print(tiene_33([1, 3, 3]))          
print(tiene_33([1, 3, 1, 3]))       
print(tiene_33([3, 1, 3, 3]))  
```

#### 7.- Escribe una función que busque numeros primos. 

```python
def es_primo(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):  # solo hasta la raíz cuadrada
        if n % i == 0:
            return False
    return True
```

```python
print(es_primo(13))
print(es_primo(2))
```

#### 8.- Juego de espías. Escribe una función que encuentre la secuencia 007 en una lista

```python
def espia_007(lista):
    codigo = [0, 0, 7]  # secuencia a encontrar
    for num in lista:
        if num == codigo[0]:
            codigo.pop(0)  # eliminamos el primero esperado
        if not codigo:  # si está vacío, encontramos toda la secuencia
            return True
    return False
```

```python
print(espia_007([1, 2, 4, 0, 0, 7, 5]))
print(espia_007([1, 7, 2, 0, 4, 5, 0]))         
```

#### 9.- Juego de BLACKJACK. Dados tres numeros enteros entre el 1 y el 11, si su suma es menor o igual que 21, devuelve su suma. Si su suma es mayor que 21 y hay un 11, resta 10 al total de la suma. Finalmente, si con el ajuste es mayor que 21, se devuelve BUST.
##### 5,6,7 =18
##### 9,9,9 = BUST
##### 9,9,11 = 19

```python
def blackjack(a, b, c):
    total = a + b + c
    
    if total <= 21:
        return total
    elif 11 in (a, b, c) and total - 10 <= 21:
        return total - 10
    else:
        return "BUST"
```

```python
print(blackjack(5, 6, 7))   
print(blackjack(9, 9, 9))    
print(blackjack(9, 9, 11))   
print(blackjack(10, 11, 11))  
```

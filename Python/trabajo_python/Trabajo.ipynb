#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
#esta es nuestra Secuencia ADN: secuencia2 (se puede ver en el primer apartado de la pregunta 5.)
```

```python
# Y estas son las enzimas que estan en un diccionario junto a sus sitios de corte:
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
```

```python
#Primero,  vamos a buscar los sitios de corte de las enzimas. Para ello, quitaré el símbolo *

enzimas_sin_asterisco = {}  #diccionario vacío donde guardaremos los sitios de corte junto a la enzima a la que corresponde (sin el * para favilitar la digestion de la secuencia)

# Recorrer el diccionario original (enzR)
for enzima, sitio in enzR.items():
    sitio_sin_asterisco = sitio.replace("*", "")        # Eliminar el '*' del sitio de corte con "replace"
    enzimas_sin_asterisco[enzima] = sitio_sin_asterisco    # Añadir la enzima y su sitio de corte sin '*' al nuevo diccionario

# Imprimir el nuevo diccionario con los sitios de corte sin el '*'
print(enzimas_sin_asterisco)
```

```python
# Con los sitios de corte limpios, ahora voy a crear una funcion para ver si hay algun sitio de corte en nuestra secuencia.  
import re

# Función para realizar la digestión
def digestión_detalles(secuencia2, enzimas_sin_asterisco):
    resultados_enzimas = {}  # Para almacenar los resultados de cada enzima
    fragmentos_resultantes_detalles = {}  # Para almacenar los fragmentos generados por cada enzima
    
    # Buscar sitios de corte para cada enzima
    for enzima, patrón in enzimas_sin_asterisco.items():
        # Buscar las posiciones de corte de la enzima
        coincidencias = [m.start() for m in re.finditer(patrón, secuencia2)]
        
        # Si la enzima corta, procesamos los fragmentos
        if coincidencias:
            # Almacenamos las posiciones de corte
            resultados_enzimas[enzima] = {"ha_cortado": True, "cant_cortes": len(coincidencias), "posiciones_corte": coincidencias}
            
            # Fragmentar la secuencia según los sitios de corte
            posiciones_de_corte = sorted(coincidencias)
            fragmentos = [secuencia2[:posiciones_de_corte[0]]]  # Primer fragmento antes del primer corte
            for i in range(1, len(posiciones_de_corte)):
                fragmentos.append(secuencia2[posiciones_de_corte[i-1]:posiciones_de_corte[i]])  # Fragmentos entre cortes
            fragmentos.append(secuencia2[posiciones_de_corte[-1]:])  # Último fragmento después del último corte
            
            # Almacenamos los fragmentos y sus tamaños
            fragmentos_resultantes_detalles[enzima] = [(len(fragmento), fragmento) for fragmento in fragmentos]
        else:
            resultados_enzimas[enzima] = {"ha_cortado": False, "cant_cortes": 0, "posiciones_corte": []}
            fragmentos_resultantes_detalles[enzima] = [(len(secuencia2), secuencia2)]  # Si no corta, la secuencia original es el único fragmento
    
    return fragmentos_resultantes_detalles, resultados_enzimas

# Ejecutar la función
fragmentos_resultantes_detalles, resultados_enzimas = digestión_detalles(secuencia2, enzimas_sin_asterisco)

# Mostrar los resultados detallados para cada enzima mediante "print"
for enzima, resultado in resultados_enzimas.items():
    print(f"\nResultados para {enzima}:")
    if resultado["ha_cortado"]:
        print(f"   Ha cortado la secuencia {resultado['cant_cortes']} veces.")
        print(f"   Las posiciones de corte son: {resultado['posiciones_corte']}")
        print(f"   Fragmentos resultantes (longitud y secuencia): {fragmentos_resultantes_detalles[enzima]}")
    else:
        print(f"   No ha cortado la secuencia.")
```

```python
#Esta sería una forma de visualizar el resultado de la digestion de nuestra secuencia de una manera más limpia (sin los fragmentos). 
#Solo se observa que enzimas han realizado cortes, la longitud de estos y cuantos.
for enzima, resultado in resultados_enzimas.items():
    print(f"\nResultados para {enzima}:")
    if resultado["ha_cortado"]:
        print(f"   Ha cortado la secuencia {resultado['cant_cortes']} veces.")
        print(f"   Las posiciones de corte son: {resultado['posiciones_corte']}")
    else:
        print(f"   No ha cortado la secuencia.")
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
# Función para obtener la secuencia complementaria inversa (cadena negativa)
def complementario_inverso(seq):
    complemento = {"A": "T", "T": "A", "C": "G", "G": "C"}
    return "".join(complemento.get(base, base) for base in reversed(seq))

# Diccionario de codones a aminoácidos
codon_to_aa = {
    'ATA': 'I', 'ATC': 'I', 'ATT': 'I', 'ATG': 'M', 'ACA': 'T', 'ACC': 'T', 'ACG': 'T', 'ACT': 'T',
    'AAC': 'N', 'AAT': 'N', 'AAA': 'K', 'AAG': 'K', 'AGC': 'S', 'AGT': 'S', 'AGA': 'R', 'AGG': 'R',
    'CTA': 'L', 'CTC': 'L', 'CTG': 'L', 'CTT': 'L', 'CCA': 'P', 'CCC': 'P', 'CCG': 'P', 'CCT': 'P',
    'CAC': 'H', 'CAT': 'H', 'CAA': 'Q', 'CAG': 'Q', 'CGA': 'R', 'CGC': 'R', 'CGG': 'R', 'CGT': 'R',
    'GTA': 'V', 'GTC': 'V', 'GTG': 'V', 'GTT': 'V', 'GCA': 'A', 'GCC': 'A', 'GCG': 'A', 'GCT': 'A',
    'GAC': 'D', 'GAT': 'D', 'GAA': 'E', 'GAG': 'E', 'GGA': 'G', 'GGC': 'G', 'GGG': 'G', 'GGT': 'G',
    'TCA': 'S', 'TCC': 'S', 'TCG': 'S', 'TCT': 'S', 'TTC': 'F', 'TTT': 'F', 'TTA': 'L', 'TTG': 'L',
    'TAC': 'Y', 'TAT': 'Y', 'TAA': 'STOP', 'TAG': 'STOP', 'TGA': 'STOP', 'TGC': 'C', 'TGT': 'C',
    'TGA': 'C', 'TGG': 'W'
}

# Función para encontrar los ORFs 
def encontrar_orfs(secuencia2):
    orfs = []  # Lista vacía para almacenar los ORFs encontrados
    for cadena in [secuencia2, complementario_inverso(secuencia2)]:
        inicio = 0
        while inicio < len(cadena):
            # Buscar codón de inicio (ATG)
            indice_inicio = cadena.find("ATG", inicio)
            if indice_inicio == -1:
                break
            # Buscar codones de stop (TAA, TAG, TGA)
            for codon_paro in ["TAA", "TAG", "TGA"]:
                indice_paro = cadena.find(codon_paro, indice_inicio + 3)
                if indice_paro != -1:
                    orfs.append(cadena[indice_inicio:indice_paro + 3])  # Añadir el ORF mediante la función append
                    inicio = indice_paro + 3
                    break
                else:
                    inicio = indice_inicio + 3  # Esto avanza a la siguiente posición
    return orfs

# Función para calcular el porcentaje de aminoácidos aromáticos (F, Y, W)
def porcentaje_aminoacidos_aromaticos(proteina):
    aromaticos = ['F', 'Y', 'W']
    cuenta_aromaticos = sum(1 for aa in proteina if aa in aromaticos)
    return (cuenta_aromaticos / len(proteina)) * 100

# Función para traducir los ORFs de ADN a proteínas (utilizando el diccionario de codones)
def traducir_orf(orf):
    proteina = ""
    for i in range(0, len(orf), 3):
        codon = orf[i:i+3]
        if codon in codon_to_aa:
            aa = codon_to_aa[codon]
            if aa != 'STOP':
                proteina += aa
            else:
                break
    return proteina

# Función para procesar los ORFs y obtener los resultados
def analizar_orfs(secuencia2):
    # Encontrar ORFs en la secuencia
    orfs = encontrar_orfs(secuencia2)
    
    # Ordenar los ORFs por longitud (de mayor a menor)
    orfs_ordenados = sorted(orfs, key=len, reverse=True)
    
    # Seleccionar los 6 ORFs más largos
    orfs_largos = orfs_ordenados[:6]
    
    # Determinar la cadena (positiva o negativa) para cada ORF
    cadena_info = []
    for orf in orfs_largos:
        if orf in secuencia2:
            cadena_info.append("positiva")
        else:
            cadena_info.append("negativa")
    
    # Traducir los ORFs a proteínas y calcular el porcentaje de aminoácidos aromáticos
    proteinas = [traducir_orf(orf) for orf in orfs_largos]
    porcentaje_aromaticos = [porcentaje_aminoacidos_aromaticos(proteina) for proteina in proteinas]
    
    # ORF más largo
    orf_mas_largo = orfs_largos[0]
    proteina_mas_larga = proteinas[0]
    porcentaje_aromaticos_max = porcentaje_aromaticos[0]
    
    # Longitud de la proteína más larga respecto a la secuencia original
    longitud_relativa = len(orf_mas_largo) / len(secuencia2) * 100
    
    # Resultados finales
    resultados = {
        "orfs_largos": orfs_largos,
        "cadena_info": cadena_info,
        "porcentaje_aromaticos": porcentaje_aromaticos,
        "orf_mas_largo": orf_mas_largo,
        "proteina_mas_larga": proteina_mas_larga,
        "porcentaje_aromaticos_max": porcentaje_aromaticos_max,
        "longitud_relativa": longitud_relativa
    }
    
    return resultados

# Ejecutar la función con la secuencia de ADN
resultados = analizar_orfs(secuencia2)

# Print para mostrar los 6 ORFs más largos y sus detalles
for i, orf in enumerate(resultados["orfs_largos"]):
    print(f"\nORF {i+1}:")
    print(f"   Secuencia: {orf}")
    print(f"   Cadena: {resultados['cadena_info'][i]}")
    print(f"   Longitud: {len(orf)}")
    print(f"   Porcentaje de aminoácidos aromáticos: {resultados['porcentaje_aromaticos'][i]:.2f}%")

# Print para mostrar el ORF más largo y su información
print(f"\nEl ORF más largo es:")
print(f"   Secuencia: {resultados['orf_mas_largo']}")
print(f"   Longitud: {len(resultados['orf_mas_largo'])}")
print(f"   Proteína: {resultados['proteina_mas_larga']}")
print(f"   Porcentaje de aminoácidos aromáticos: {resultados['porcentaje_aromaticos_max']:.2f}%")
print(f"   Longitud relativa a la secuencia original: {resultados['longitud_relativa']:.2f}%")
```

#### 12.- Crea un archivo mediante código que se llame _tabla-nombres.txt_ y que esté formado por 4 columnas y 4 filas con la siguiente formación (1,5 ptos):
#### Primera columna: Nombre
#### Segunda columna: Apellidos (2 apellidos)
#### Tercera columna: Edad
#### Cuarta columna: Mascotas (nombre de la mascota si tienen)

#### ¿Qué pasa si abres el archivo desde Jupyter y has metido acentos? ¿Puedes solucionarlo?

```python
filas = [
    ["Nombre", "Apellidos", "Edad", "Mascotas"],
    ["Ane", "Ruiz castellano", "63", "Luna"],
    ["Juan", "Aguirre delgado", "46", "Coco"],
    ["Leire", "Palmero castañeda", "18", "bruno"]
]
 
# Crear y escribir en el archivo
with open("tabla-nombres.txt", "w", encoding="utf-8") as archivo:
    for fila in filas:
        # Unimos las columnas con tabulaciones
        archivo.write("\t".join(fila) + "\n")

# Mostramos el contenido como se vería en el archivo
for fila in filas:
    print("\t".join(fila))
```

#### 13.- Crea un diccionario que contenga las siguientes parejas. Obtén el listado de keys() y guárdalo en una lista. Añade un par de elementos más y después convierte esa lista en un set (1 pto).

#### Rattus = novergicus, rattus
#### Glaucidium = brasilianum, hoskinsii, californicum, minutissimum
#### Echinocactus = grusonii
#### Opuntia = 1


```python
 
plantas_animales = {
    "Rattus": ["novergicus", "rattus"],
    "Glaucidium": ["brasilianum", "hoskinsii", "californicum", "minutissimum"],
    "Echinocactus": ["grusonii"],
    "Opuntia": [1]
}

# Obtener el listado de keys y guardarlo en una lista
keys_lista = list(plantas_animales.keys())
print("Lista de keys:", keys_lista)

# Añadir un par de elementos más al diccionario
plantas_animales["Ficus"] = ["elastica"]
plantas_animales["Ailuropoda"] = ["melanoleuca"]

# Convertir la lista de keys en un set
keys_set = set(list(plantas_animales.keys()))
print("Set de keys:", keys_set)
```
